#비선형 자료구조

(1) 트리: 계층적, 사이클 안됨

예시) 가계도

이진트리

```python
tree={'A':['B','C'],'B':['D','E'],'C':['F','G']}

# 딕셔너리에서 아이템즈써서 키와 벨류를 한 쌍으로
for i,j in tree.items():
  print(f'{i}의 자식은 {j}입니다')
```

```python
tree={'A':['B','C'],'B':['D','E'],'C':['F','G']}

for i in tree.keys():
  # 해당하는 값들을 ,로 묶는 방법인 join
  c=','.join(tree[i])
  print(f'{i}의 자식은 {c}입니다')
```

```python
# 아이템즈와 조인을 같이 써보기

tree={'A':['B','C'],'B':['D','E'], 'C':['F']}

for i,j in tree.items():
  c=','.join(j)
  print(f'{i}의 자식은 {c}입니다')
```

```python
# 트리를 딕셔너리로 연습

tree={
    'A':['B','C'],
    'B':['D','E'],
    'C':['F'],
    'D':[],
    'E':[],
    'F':[]
}

import networkx as nx
import matplotlib.pyplot as plt


G=nx.Graph(tree) # 그래프 생성
pos=nx.spring_layout(G)
nx.draw(G,pos,with_labels=True,node_color= 'r',edge_color='y',linewidths=1)
plt.show()
```

```python
# 트리를 딕셔너리로 연습

tree={
    'A':['B','C'],
    'B':['D','E'],
    'C':['F'],
    'D':[],
    'E':[],
    'F':[]
}

import networkx as nx
import matplotlib.pyplot as plt


G=nx.DiGraph(tree) # 방향이 있는 그래프 생성
pos=nx.spring_layout(G)
nx.draw(G,pos,with_labels=True,node_color= 'r',edge_color='y',linewidths=1)
plt.show()
```

```python
# 딕셔너리가 아니라 리스트로 트리 표현하기

#                A
#        B                C
#    D       E         F       G

tree=['A',['B',['D','E']],['C',['F','G']]]

print(f'전체 트리는 {tree}이다')
print(f'루트 노드는 {tree[0]}이다')
print(f'루트노드의 왼쪽 자식노드는 {tree[1][0]}이다')  # [1][0]이 포인트  #B
print(f'루트노드의 오른쪽 자식노드는 {tree[2][0]}이다')  #C
print(f'B의 왼쪽 자식노드는 {tree[1][1][0]}이다')   #D
print(f'B의 오른쪽 자식노드는 {tree[1][1][1]}이다')  #E
print(f'C의 왼쪽 자식노드는 {tree[2][1][0]}이다')   #F
print(f'C의 오른쪽 자식노드는 {tree[2][1][1]}이다')  #G
```

#비선형 자료구조

(2) 그래프: 사이클 생성가능, 부모자식개념 없음

예시) 지하철 노선도

그래프에 속한 것이 트리

```python
# 인접행렬을 이용한 그래프의 표현
# 행: 출발
# 열: 도착
# 요소: 0=인접(간선으로 연결)하지 않음, 1=인접함


size=4 # 전역변수

def graph(size):
    graph=[]
    for i in range(size):
        graph.append([0]*size)  # 기본값으로 모든 요소를 0으로 세팅해둠
    return graph
G1=graph(size)
print("기본")
print(G1)
print()

# 0으로 세팅된 모든요소들 중 일부를 1로 수정
G1[0][1]=1  #A와 B는 간선으로 연결됐다는 의미
G1[0][2]=1
G1[0][3]=1

G1[1][0]=1
G1[1][2]=1

G1[2][0]=1
G1[2][1]=1
G1[2][3]=1

G1[3][0]=1
G1[3][2]=1

print("첫번째 방법")
for i in range(size):
  print(f'{i+1}번째 행: {G1[i]}')

print()
print("두번째 방법")
for row in range(size):
  for col in range(size):
    print(G1[row][col],end=' ')
  print()
```

```python
G3=graph(size)
print(G3)
G3[0][1]=1
G3[0][2]=1
G3[3][0]=1
G3[3][2]=1

for i in range(size):
  print(f'{i+1}번째 행: {G3[i]}')
```
